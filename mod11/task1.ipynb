Без использования numpy

```python
def mult(a, b):
    result = None
    rows_a = len(a)
    cols_a = len(a[0])
    rows_b = len(b)
    cols_b = len(b[0])

    # Проверка равенства столбцов
    if cols_a != rows_b:
        raise print("Количество столбцов в первой матрице должно быть равно количеству строк во второй")

    # Создаем результирующую матрицу с нулями
    result = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

    # Перемножаем матрицы
    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result[i][j] += a[i][k] * b[k][j]
    return result
```

C использованием numpy

```python
def np_mult(a, b):
    result = None
    result = np.dot(a, b)
    return result
```

```python
%%time
# засечем время работы функции без NumPy
M1 = mult(a, b)
```

```python
%%time
# засечем время работы функции с NumPy
M2 = np_mult(a, b)
```

```python
# проверим корректность
np.abs(np.array(M1) - M2).sum()
```

### Задание 3
Напишите функцию, которая находим сумму четных диагональныех элементов квадратной матрицы. Если таких элементов нет, то вывести `0`.

```python
# функция, решающая задачу с помощью NumPy
def np_diag_2k(a):
  res = None
  # Извлечение диагональных элементов
  diag_elements = np.diag(a)
  # Фильтрация четных элементов
  even_elements = diag_elements[diag_elements % 2 == 0]
  # Суммирование четных элементов
  res = even_elements.sum()
  return res
```

```python
# зададим некоторую квадратную матрицу
a = np.random.randint(1, 10, size=(5, 5))
a
```

```python
%%time
# засечем время работы функции с NumPy
np_diag_2k(a)
```

### Задание 4
На вход дан двумерный массив $X$. Напишите функцию, которая для каждой строчки $x = (x_1, x_2, \ldots, x_n)$ массива $X$  строит строчку $s = (s_1, s_2, \ldots, s_n)$, где $s_k=\frac{x_1+...+x_k}{k}$, а затем выдаёт массив из построенных строчек. Используйте библиотеку numpy (вам поможет функция cumsum). Выходом функции должен быть двумерный массив той же формы, что и $X$.

```python
# функция, решающая задачу с помощью NumPy
def np_sec_av(*A):
    # Извлечение массива из кортежа
    A = A[0]
    # Проверка размерности массива и преобразование в двумерный при необходимости
    if A.ndim == 1:
        A = A.reshape(1, -1)
    # Вычисление кумулятивной суммы вдоль строк
    cum_sum = np.cumsum(A, axis=1)
    # Получение количества элементов для деления
    divisors = np.arange(1, A.shape[1] + 1)
    # Вычисление средних значений
    result = cum_sum / divisors
    return result
```

```python
# зададим некоторую последовательность и проверим ее на ваших функциях.
# функция должна работать ~ в 50 раз быстрее, чем без Numpy
A = sps.uniform.rvs(size=10**3)
print(A)

%time S2 = np_sec_av(A)
```

### Задание 5
Написать функцию для кодирование массива (Run-length encoding). Все подряд повторения элементов функция сжимает в один элемент и считает количество повторений. Функция возвращает кортеж из двух векторов одинаковой длины. Первый содержит элементы, а второй — сколько раз их нужно повторить.

Пример: encode(np.array([1, 2, 2, 3, 3, 1, 1, 5, 5, 2, 3, 3])) = (np.array[1, 2, 3, 1, 5, 2, 3]), np.array[1, 2, 2, 2, 2, 1, 2])

```python
# функция, решающая задачу с помощью NumPy
def np_encode(a):
    result = None
    # Находим индексы, где значение изменяется
    change_indices = np.where(np.diff(a, prepend=a[0]-1, append=a[-1]+1) != 0)[0]
    # Элементы - это значения в индексах изменений
    elements = a[change_indices[:-1]]
    # Частоты - разница между последовательными индексами изменений
    frequencies = np.diff(change_indices)
    result = (elements, frequencies)
    return result
```

### Задание 6

Пусть задан некоторый одномерный массив $X$. Надо построить новый массив, где все элементы с нечетными индексами требуется заменить на число $a$ (если оно не указано, то на 1). Все элементы с четными индексами нужно возвести в куб. Затем записать элементы в обратном порядке относительно их позиций. Массив $X$ при этом должен остаться *без изменений*. В конце требуется слить массив X с преобразованным X и вывести в обратном порядке.  

Пример:  
X = [100, 200, 300, 400, 500] -> [100, a, 300, a, 500] -> [500^3, a, 300^3, a, 100^3] -> склеиваем -> [100, 200, 300, 400, 500, 500^3, a, 300^3, a, 100^3]

```python
# функция, решающая задачу с помощью NumPy
from copy import copy

def np_transformation(X, a=1):
    """
    param X: array
    """
    result = None
    # Создаем копию массива X
    Y = copy(X)
    # Замена элементов с нечетными индексами на число a
    Y[1::2] = a
    # Возведение элементов с четными индексами в куб
    Y[::2] = np.power(Y[::2], 3)
    # Переворачиваем массив Y
    Y = Y[::-1]
    # Склеиваем X и Y, затем переворачиваем
    result = np.concatenate((X, Y))[::-1]
    return result
```

```python
X = np.array([i for i in range(1, 10, 2)])
# функция должна работать ~ в 20 раз быстрее, чем без Numpy
%time S2 = np_transformation(X, 5)
```
